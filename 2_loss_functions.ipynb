```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define true values (y_true) and model predictions (y_pred)
y_true_mse = np.array([3.0, -0.5, 2.0, 7.0])
y_pred_mse = np.array([2.5, 0.0, 2.0, 8.0])

y_true_cce = np.array([[0, 1, 0], [0, 0, 1]])  # One-hot encoded true labels
y_pred_cce = np.array([[0.05, 0.85, 0.10], [0.10, 0.20, 0.70]])  # Predicted probabilities

# Step 2: Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE) Loss
mse_loss_fn = tf.keras.losses.MeanSquaredError()
cce_loss_fn = tf.keras.losses.CategoricalCrossentropy()

mse_loss = mse_loss_fn(y_true_mse, y_pred_mse).numpy()
cce_loss = cce_loss_fn(y_true_cce, y_pred_cce).numpy()

print(f"Initial MSE Loss: {mse_loss}")
print(f"Initial Categorical Cross-Entropy Loss: {cce_loss}")

# Step 3: Modify predictions slightly and compute new loss values
y_pred_mse_modified = np.array([2.7, -0.2, 2.1, 7.5])
y_pred_cce_modified = np.array([[0.10, 0.80, 0.10], [0.15, 0.25, 0.60]])

mse_loss_modified = mse_loss_fn(y_true_mse, y_pred_mse_modified).numpy()
cce_loss_modified = cce_loss_fn(y_true_cce, y_pred_cce_modified).numpy()

print(f"Modified MSE Loss: {mse_loss_modified}")
print(f"Modified Categorical Cross-Entropy Loss: {cce_loss_modified}")

# Step 4: Plot loss function values using Matplotlib
labels = ['Initial', 'Modified']
mse_values = [mse_loss, mse_loss_modified]
cce_values = [cce_loss, cce_loss_modified]

x = np.arange(len(labels))  # Label positions
width = 0.3  # Bar width

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, mse_values, width, label='MSE Loss', color='b')
plt.bar(x + width/2, cce_values, width, label='CCE Loss', color='r')

plt.xlabel('Prediction Type')
plt.ylabel('Loss Value')
plt.title('Comparison of MSE and Categorical Cross-Entropy Loss')
plt.xticks(ticks=x, labels=labels)
plt.legend()
plt.show()
```
